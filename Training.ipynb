### Viene utilizzato il label encoder per rendere i valori in tipi numerici, invece di tenerli nel tipo originale. Questo passaggio servir√† soprattutto per analizzare le heatmap e i barchart che verrano costruiti successivamente 

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
train['ssc_b'] = labelEncoder.fit_transform(train['ssc_b'])
train['hsc_b'] = labelEncoder.fit_transform(train['hsc_b'])
train['hsc_s'] = labelEncoder.fit_transform(train['hsc_s'])
train['degree_t'] = labelEncoder.fit_transform(train['degree_t'])
train['specialisation'] = labelEncoder.fit_transform(train['specialisation'])
train['workex'] = labelEncoder.fit_transform(train['workex'])
train['status'] = labelEncoder.fit_transform(train['status'])

test['ssc_b'] = labelEncoder.fit_transform(test['ssc_b'])
test['hsc_b'] = labelEncoder.fit_transform(test['hsc_b'])
test['hsc_s'] = labelEncoder.fit_transform(test['hsc_s'])
test['degree_t'] = labelEncoder.fit_transform(test['degree_t'])
test['specialisation'] = labelEncoder.fit_transform(test['specialisation'])
test['workex'] = labelEncoder.fit_transform(test['workex'])
test['status'] = labelEncoder.fit_transform(test['status'])
```

### Si utilizzano le heat map per verificare le correlazioni tra le varie features e la colonna 'status'. Una correlazione molto bassa ci suggerisce di eliminare la colonna. Utilizziamo le heat map anche per verificare eventuali correlazioni tra feature

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
def heat_map():
    correlation = train.corr()
    plt.figure(figsize=(20,20))
    sns.heatmap(correlation, square=True,annot=True,cmap='cubehelix')
    plt.title('Correlation between different features')
```

```python
heat_map()
```

### Costruzione e Fit dei modelli: Decision Tree e Random Forest

```python
y_train = train["status"].copy()
X_train = train.drop("status",axis=1).copy()

y_test = test["status"].copy()
X_test = test.drop("status",axis=1).copy()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
trainedtree = model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Decision Tree Classifier")
print("accuracy score:", accuracy_score(y_test, pred)*100)
print("model score:", model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
randomForest_model = RandomForestClassifier(n_estimators=100)
randomForest_model.fit(X_train,y_train)
y_pred = randomForest_model.predict(X_test)
print("Random forest")
print("accuracy score:", accuracy_score(y_test, y_pred))
print("model score:", randomForest_model.score(X_test, y_test))
```
